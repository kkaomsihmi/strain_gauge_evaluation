```python
#完成版
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

def plot_and_save_scatter(csv_file_path, fig_title):
    # CSVファイルのパスを指定
    csv_file_path = Path(csv_file_path)

    # CSVファイルを読み込み、DataFrameに変換
    df = pd.read_csv(csv_file_path, encoding='shift_jis')  # 'shift_jis' は適切なエンコーディングに置き換えてください

    # DataFrameの内容を表示（デバッグ用）
    print(df)

    # グラフを作成
    fig, ax = plt.subplots()
    ax.scatter(x=df["日時(μs)"], y=df["(NR-1)ST-CH01"], s=1, label=df.columns[2])
    ax.scatter(x=df["日時(μs)"], y=df["(NR-1)ST-CH02"], s=1, color="r", label=df.columns[3])

    # 軸ラベルの設定
    ax.set_xlabel('time')
    ax.set_ylabel('strain_gauge')

    # 凡例の設定
    ax.legend()
    ax.grid(True)
    ax.legend(loc="upper left", bbox_to_anchor=(1.02, 1.0), borderaxespad=0)

    ax.set_title(fig_title)

    # グラフを表示
    plt.show()

    # グラフをPNGファイルとして保存
    output_file_path = csv_file_path.with_suffix('.png')
    fig.savefig(output_file_path, bbox_inches='tight')
    print(f"グラフが保存されました: {output_file_path}")
```

```python
#完成品(20240523)
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib_fontja  # Matplotlibで日本語を使用可能にする
from pathlib import Path
from openpyxl.drawing.image import Image
import tkinter as tk
from tkinter import filedialog
import sys

def select_csv_files():
    # ファイル選択ダイアログを開く
    file_paths = filedialog.askopenfilenames(
        title="CSVファイルを選択してください",
        filetypes=[("CSVファイル", "*.csv;*.CSV"), ("すべてのファイル", "*.*")]
    )
    
    # ファイルが選択されていない場合
    if not file_paths:
        messagebox.showerror("エラー", "ファイルが選択されていません。")
        return None

    # 全ての選択されたファイルがCSVファイルであることを確認
    for file_path in file_paths:
        if not file_path.lower().endswith('.csv'):
            messagebox.showerror("エラー", "選択されたファイルの中にCSVファイルではないものがあります。")
            return None

    return list(file_paths)

def extract_file_names(file_paths):
    # ファイルパスからファイル名を抽出
    file_names = [Path(file_path).name for file_path in file_paths]
    return list(file_names)

def plot_and_save_scatter(csv_file_path, fig_title):
    csv_file_path = Path(csv_file_path)
    # CSVファイルを読み込み、DataFrameに変換
    df = pd.read_csv(csv_file_path, encoding='shift_jis')  # 'shift_jis' は適切なエンコーディングに置き換えてください

    dfs.append(df)
    # DataFrameの内容を表示（デバッグ用）
    print(df)

    # グラフを作成
    fig, ax = plt.subplots()
    ax.scatter(x=df["日時(μs)"], y=df["(NR-1)ST-CH01"], s=1, label=df.columns[2])
    ax.scatter(x=df["日時(μs)"], y=df["(NR-1)ST-CH02"], s=1, color="r", label=df.columns[3])

    # 軸ラベルの設定
    ax.set_xlabel('time')
    ax.set_ylabel('strain_gauge')

    # 凡例の設定
    ax.legend()
    ax.grid(True)
    ax.legend(loc="upper left", bbox_to_anchor=(1.02, 1.0), borderaxespad=0)

    ax.set_title(fig_title)

    # グラフを表示
    plt.show()

    # グラフをPNGファイルとして保存
    output_file_path = csv_file_path.with_suffix('.png')
    fig.savefig(output_file_path, bbox_inches='tight')
    print(f"グラフが保存されました: {output_file_path}")

#exc
csv_file_list=select_csv_files()
dfs=[]

for i in range(0,len(csv_file_list)):
    csv_list=extract_file_names(csv_file_list)

for i in range(0,len(csv_file_list)):
    plot_and_save_scatter(csv_file_list[i], csv_list[i])

max_strain_gauge = pd.Series([dfs[i]["(NR-1)ST-CH01"].max() for i in range(0,len(csv_file_list))], index=csv_list)
#各dfの最大値を抽出(CH01)
max_df=pd.DataFrame(max_strain_gauge, columns=["max"])
#max_strain_gaugeをdataframe化
min_strain_gauge = pd.Series([dfs[i]["(NR-1)ST-CH02"].min() for i in range(0,len(csv_file_list))], index=csv_list)
#各dfの最小値を抽出(CH02)
min_df=pd.DataFrame(min_strain_gauge,columns=["min"])
#min_strain_gaugeをdataframe化

concat_df=pd.concat([min_df, max_df], axis=1)#最大・小値を結合
concat_df_describe=pd.concat([min_df.describe(), max_df.describe()], axis=1)

# データフレームをExcelファイルに出力
#output_file=Path(Path.cwd(),"output.xlsx")
with pd.ExcelWriter("output.xlsx") as writer:
    concat_df.to_excel(writer, index=True,sheet_name="data")
    concat_df_describe.to_excel(writer, index=True,sheet_name="describe")
```
